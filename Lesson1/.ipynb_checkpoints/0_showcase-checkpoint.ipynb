# Reading input files with Pandas
For ROOT files, you'll need root_numpy, which you can find here: https://github.com/rootpy/root_numpy

You can actually even convert directly to pandas without need for numpy with root_pandas, here: https://github.com/ibab/root_pandas

```python
def Pandafy(fileName, tree):
    df = pd.DataFrame(rnp.root2array(fileName,tree))
    return df

df = Pandafy('Data/tree.root','Tree')
print "TFile pandafied"
```
