```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('Data/randomData.csv')
plt.step(df['X'],df['Y'])
plt.step(df['X'],df['Y'].cumsum()/30.)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('A plot')
plt.ylim(0,3)
plt.grid()
plt.legend()
plt.show()
```
