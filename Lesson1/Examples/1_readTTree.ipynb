# Read in with PyROOT
Very simple, you won't need anything beside PyROOT, but it's nothing extremely fancy.

```python
import ROOT

rFile = ROOT.TFile('../Data/tree.root')
rTree = rFile.Get('Tree')

for entry in range(rTree.GetEntries()):
    rTree.GetEntry(entry)
    for length in rTree.TrackLength:
        print length
```

# Read in with root_numpy + pandas

As you can see you'll need these three modules.

Very fast, and you get everything on pandas dataframe.

I've written a small function that wraps it up, but you don't have to use it.

```python
import pandas as pd
import numpy as np
import root_numpy as rnp

def Pandafy(fileName, tree):
    df = pd.DataFrame(rnp.root2array(fileName,tree))
    return df

df = Pandafy('../Data/tree.root','Tree')

df
```

# Save it, and read it back (pickle)

```python
df.to_pickle('../Data/tree.pickle')
df_frompickle = pd.read_pickle('../Data/tree.pickle')
print df_frompickle['TrackLength'].at[0][0]
df_frompickle
```

# Save it, and read it back (hdf5)

```python
df.to_hdf('../Data/tree.hdf','tree')
df_fromhdf = pd.read_hdf('../Data/tree.hdf','tree')
print df_fromhdf['TrackLength'].at[0][0]
df_fromhdf
```
