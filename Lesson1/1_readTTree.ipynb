# Read in with PyROOT
Very simple, you won't need anything beside PyROOT, but it's nothing extremely fancy.

```python
import ROOT

rFile = ROOT.TFile('tree.root')
rTree = rFile.Get('Tree')

for entry in range(2):
    rTree.GetEntry(entry)
    for length in rTree.TrackLength:
        print length
```

# Read in with root_numpy + pandas

As you can see you'll need these three modules.

Very fast, and you get everything on pandas dataframe.

I've written a small function that wraps it up, but you don't have to use it.

```python
import pandas as pd
import numpy as np
import root_numpy as rnp

def Pandafy(fileName, tree):
    df = pd.DataFrame(rnp.root2array(fileName,tree))
    return df

df = Pandafy('tree.root','Tree')

df
```
