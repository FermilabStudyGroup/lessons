# Widgets
Why not combining matplotlib with widgets for easier exploration and plot manipulation?

For more info, I've made an example here: https://github.com/sdporzio/codingExamples/blob/master/Jupyter/widgets_3dPlots.ipynb

You can find instructions here:
https://ipywidgets.readthedocs.io/en/latest/

```python
from ipywidgets import interact, interactive, fixed, widgets

def plot(Pdg=211,ratio=50):
    fig = plt.figure(figsize=figsize)
    
    # Apply cuts
    cut = (df.Pz/np.sqrt(df.Px**2 + df.Py**2) > ratio)
    cutDf = df[cut]
    n,hist,_ = plt.hist(df[df.Pdg==Pdg]['Pz'],100,label='PreCut')
    n,hist,_ = plt.hist(cutDf[cutDf.Pdg==Pdg]['Pz'],100,label='PostCut')
    plt.xlabel('Momentum [eV]')
    plt.legend()
    plt.grid()
    plt.show()
    
w_Pdg = {'Kaon (321)':321,'Pion (211)':211}
w_Ratio = (0,100,5)
```

```python
interact(plot,Pdg=w_Pdg,ratio=w_Ratio)
```

#### You can do the same for easier visualization, especially for 3d plots
The widgets can take care of any variable

```python
from ipywidgets import interact, interactive, fixed, widgets
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets

iris = datasets.load_iris()
data = iris.data
test = iris.target
data1 = data[np.where(test==0)]
data2 = data[np.where(test==1)]
data3 = data[np.where(test==2)]

def look(elevation=10,azimuth=75):
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(data1[:,0],data1[:,1],data1[:,2],s=20,c='r',label='Setosa')
    ax.scatter(data2[:,0],data2[:,1],data2[:,2],s=20,c='b',label='Vergicolour')
    ax.scatter(data3[:,0],data3[:,1],data3[:,2],s=20,c='g',label='Virginica')
    ax.view_init(elev=elevation, azim=azimuth)
    plt.legend()
    plt.show()
    
w_elev = (0,90,5)
w_azim = (0,90,5)
```

```python
interact(look, elevation=w_elev, azimuth=w_azim)
```
